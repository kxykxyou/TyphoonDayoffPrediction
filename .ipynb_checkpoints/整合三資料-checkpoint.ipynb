```python
'''
df_observations資料處理流程
'''
## 生成完整日期
df_observations['ObsTime'].astype('string')
df_observations['ObsDate'] = df_observations['file_name'].map(lambda x: x[:7])+'-'+df_observations['ObsTime'].map('{:02}'.format)

### 從df_typhoon_flattened獲取警報日期
alert_dates = df_typhoon_flattened['alert_dates'].unique()
df_date = df_observations[df_observations['ObsDate'].isin(alert_dates)]


### 建立完整日期、ID、站名等資訊
df_date['ID'] = df_date['file_name'].map(lambda x: x[8:14])
df_date['stname'] = df_date['file_name'].map(lambda x: x[15:])

### reindex
df_date.reset_index(drop = True, inplace = True)

### ['...', 'x', '&', '/']取代為NA
### 'T'取代為0.05
df_date.replace(['...', 'X', '&', '/'], np.nan, inplace = True)
df_date.replace(['T'], 0.05, inplace = True)

### reorder columns
col_order = ['ObsDate', 'ID', 'stname',]
Time_col = []
for col in df_date.columns:
    if 'Time' in col:
        Time_col.append(col)
    else:
        if col not in col_order:
            col_order.append(col)
col_order.extend(Time_col)
df_date = df_date[col_order]

### drop不需要的columns，包含max time
col_drop = ['ObsTime', 'file_name']
df_date.drop(col_drop, axis = 1, inplace = True)
df_1 = df_date
df_1.drop(df_1.columns[-9:], axis = 1, inplace = True)

'''
新dataframe = df_1
刪除非警報日期觀測數據、獲得ID、站名、drop file_name、drop ObsTime、重新整理index
'''
```

```python
'''
reorder/drop col、rename、dtypes轉換、drop NA > 50%的
分支成：
df_3：測站為單位
df_4：鄉鎮市為單位
df_5：鄉鎮市為單位
col_clean --> wind_transform --> set_dtypes --> drop_na_ratio --> save

未完成：
township_mapping
county_mapping
'''

# ### 將不會拿進去model train的資訊以str_code串起來作為id對應
# ### 包含['ObsDate', 'ID', 'stname', 'Region_region','颱風']
def col_clean(df):
    '''
    保留並重新排列有用的columns
    重新命名部分columns
    '''
    col_keep = [
        'ObsDate',
        'ID',
        'stname',
        'Region',
        'County',
        'township',
        'altitude',
        'longitude',
        'ty_id',
        'ty_name_ch',
        'ty_name',
        'route',
        'intensity',
        'hpa',
        'wind_speed',
        '7_radius',
        '10_radius',
        'alert_num',
        'born_spotN',
        'born_spotE',
        'StnPres',
        'SeaPres',
        'StnPresMax',
        'StnPresMin',
        'Temperature',
        'T Max',
        'T Min',
        'Td dew point',
        'RH',
        'RHMin',
        'WS',
        'WD',
        'WSGust',
        'WDGust',
        'Precp',
        'PrecpHour',
        'PrecpMax10',
        'PrecpMax60',
        'SunShine',
        'SunShineRate',
        'GloblRad',
        'VisbMean',
        'EvapA',
        'UVI Max',
        'Cloud Amount',
        'day_off',
    ]
    df = df[col_keep]

    renames = {
        'ID': 'StnID',
        'stname': 'StnName',
        'ty_id': 'TyID',
        'ty_name_ch': 'TyNameCh',
        'ty_name': 'TyName',
        'wind_speed': 'TyWS',
        'day_off': 'Dayoff',
        'township':'Township',
    }

    df.rename(renames, axis = 1, inplace = True)
    return df

def township_mapping(df):
    df.groupby(['region_mapping']).agg('mean')
    
    return df

def county_mapping(df):
    df.groupby(['region_mapping']).agg('mean')
    
    return df

# ### aggregation做平均，風向做完後再轉
# ### df_2.groupby(['region_mapping']).agg('mean')


def wind_transform(df):
    '''
    風向mapping轉換
    轉換公式：(WD+11.5)//22.5，看屬於哪個區段
    '''

    wind_mapping = {
        '0': 'N',
        '1': 'NNE',
        '2': 'NE',
        '3': 'ENE',
        '4': 'E',
        '5': 'ESE',
        '6': 'SE',
        '7': 'SSE',
        '8': 'S',
        '9': 'SSW',
        '10': 'SW',
        '11': 'WSW',
        '12': 'W',
        '13': 'WNW',
        '14': 'NW',
        '15': 'NNW',
        '16': 'N'
    }
    df['WD'] = df['WD'].map(lambda x: wind_mapping[str(int((float(x)+11.25)//22.5))], na_action='ignore')
    df['WDGust'] = df['WDGust'].map(lambda x: wind_mapping[str(int((float(x)+11.25)//22.5))], na_action='ignore')
    return df

def set_dtype(df):
    '''
    要先做完wind transform才行
    '''
    dtypes = {
        'ObsDate': 'datetime64',
        'StnID': 'object',
        'StnName': 'object',
        'Region': 'object',
        'County': 'object',
        'Township': 'object',
        'altitude': 'float64',
        'longitude': 'float64',
        'TyID': 'object',
        'TyNameCh': 'object',
        'TyName': 'object',
        'route': 'category',
        'intensity': 'category',
        'hpa': 'float64',
        'TyWS': 'float64',
        '7_radius': 'float64',
        '10_radius': 'float64',
        'alert_num': 'float64',
        'born_spotN': 'float64',
        'born_spotE': 'float64',
        'StnPres': 'float64',
        'SeaPres': 'float64',
        'StnPresMax': 'float64',
        'StnPresMin': 'float64',
        'Temperature': 'float64',
        'T Max': 'float64',
        'T Min': 'float64',
        'Td dew point': 'float64',
        'RH': 'float64',
        'RHMin': 'float64',
        'WS': 'float64',
        'WD': 'category',
        'WSGust': 'float64',
        'WDGust': 'category',
        'Precp': 'float64',
        'PrecpHour': 'float64',
        'PrecpMax10': 'float64',
        'PrecpMax60': 'float64',
        'SunShine': 'float64',
        'SunShineRate': 'float64',
        'GloblRad': 'float64',
        'VisbMean': 'float64',
        'EvapA': 'float64',
        'UVI Max': 'float64',
        'Cloud Amount': 'float64',
        'Dayoff':'category',
    }
    df = df.astype(dtypes)
    return df

def drop_na_ratio(df, ratio = 0.5):
    '''
    drop掉所有觀測數值當中NA比例大於ratio的資料
    '''
    drop_idx = []
    for idx in df.index:  
        if (dfdf.loc[idx, numeric_col].isna().sum())/len(df.columns) > ratio :
            drop_idx.append(idx)
    df.drop(drop_idx, axis = 0, inplace = True)
    return df

def save(df, path):
    df.to_csv(path)
    print('dataframe saved in: ', path)
```
